```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import OrdinalEncoder
```

```python
iris=sns.load_dataset('iris')
```

```python
def ordinal(val):
    if val=='setosa':
        return 1
    elif val=='versicolor':
        return 2
    else:
        return 3
```

```python
iris['Encoded']=iris.species.apply(ordinal)
```

```python
iris
```

```python
X=iris.drop(['species','Encoded'],axis=1)
y_tf=iris.Encoded
```

```python
x_transformed=PCA(n_components=3).fit_transform(X)
x_transformed
```

```python
x=x_transformed[:,0]
y=x_transformed[:,1]
z=x_transformed[:,1]
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=y_tf)
plt.show()
```
